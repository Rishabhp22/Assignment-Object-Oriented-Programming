🔹 Definition

👉 Abstract methods are methods declared in a base class but do not have any implementation. They must be implemented by the child class.

⚔️ Abstract vs Normal Method
Feature	Abstract Method	Normal Method
Body	❌ No body	✅ Has body
Implementation	Child class must implement	Already implemented
Object creation	Cannot create object of abstract class	Can create object
Purpose	Define structure	Perform task

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass   # no implementation

class Dog(Animal):
    def sound(self):
        print("Bark")
```

```python
# a = Animal() ❌ ERROR
d = Dog()      # ✅ works
d.sound()
```
