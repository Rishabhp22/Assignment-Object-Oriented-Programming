👉 Abstraction can be used to create a common interface by defining an abstract base class that contains abstract methods. All related classes inherit from this base class and provide their own implementation of those methods.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        print("Area of circle")

class Square(Shape):
    def area(self):
        print("Area of square")
```

```python
c = Circle()
c.area()

s = Square()
s.area()
```
