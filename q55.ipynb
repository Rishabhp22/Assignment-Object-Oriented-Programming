```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before execution")
        
        result = func(*args, **kwargs)   # passing arguments
        
        print("After execution")
        return result
    return wrapper


@my_decorator
def add(a, b):
    return a + b


@my_decorator
def greet(name):
    print(f"Hello {name}")


# Testing
print(add(5, 3))
greet("Rishabh")
```
