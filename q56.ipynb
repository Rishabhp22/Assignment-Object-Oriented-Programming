```python
from functools import wraps

def my_decorator(func):
    @wraps(func)
    def wrapper():
        print("Before execution")
        func()
        print("After execution")
    return wrapper


@my_decorator
def greet():
    """This is greet function"""
    print("Hello")


greet()

print(greet.__name__)
print(greet.__doc__)
```
