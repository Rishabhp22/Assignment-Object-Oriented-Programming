What is __str__ ?

👉 __str__ is a special method (dunder method) in Python
👉 It is used to define how an object should be represented as a string

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student name is: {self.name}"

s = Student("Rishabh Poddar")
print(s)
```
