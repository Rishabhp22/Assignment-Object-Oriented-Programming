```python
def palindrome_num(n):
    for num in range(1,n+1):
        original = num
        reversed_num = 0

        while num>0:
            digit = num%10
            reversed_num = reversed_num*10+digit
            num//=10

        if original == reversed_num:
            yield original
```

```python
for num in palindrome_num(100):
    print(num)
```
