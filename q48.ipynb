Yes, encapsulation can be bypassed in Python.

Because:
👉 Python does NOT have true private variables
👉 It uses name mangling

⚙️ How it is bypassed

👉 Private variable:

```python
self.__balance
```

👉 Python converts it to:

```python
_ClassName__balance
```

Example

```python
class Bank:
    def __init__(self):
        self.__money = 1000

b = Bank()

# Direct access ❌
# print(b.__money)

# Bypassing encapsulation ✅
print(b._Bank__money)
```
