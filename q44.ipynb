```python
class Animal:
    def sound(self):
        print("Animal sound")
        
class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

class Cow(Animal):
    def sound(self):
        print("Moo")

animals = [Dog(), Cat(), Cow()]

for a in animals:
    a.sound()
```
