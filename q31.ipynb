Definition (clear)

👉 Method overriding is a feature in OOP where a child class provides a new implementation of a method that is already defined in the parent class.

🔥 Key idea

👉 Same method name
👉 Same parameters
👉 Different behavior (in child class)

```python
class Student:
    def Class(self):
        print("Class 10")

class Student2:
    def Class(self):
        print("Class 12")

c = Student2()
c.Class()
```
