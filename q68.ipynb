What is a Setter?

👉 A setter is a method used to update (modify) the value of a private variable in a controlled way.

```python
class Student:
    def __init__(self, name):
        self.__name = name   # private variable

    def set_name(self, name):
        self.__name = name

    def get_name(self):
        return self.__name


s = Student("Rishabh")

s.set_name("Rahul")
print(s.get_name())
```
