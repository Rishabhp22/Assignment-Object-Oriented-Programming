```python
class Person:
    def __init__(self, name, email):
        self.__name = name
        self.__email = email

    def set_email(self, email):
        self.__email = email

    def get_email(self, email):
        return self.__email
```

```python
p = Person("Rishabh", "hello@gmail.com")
p.get_email("Rishabh")
p.set_email('he@yahooo.com')
p.get_email("Rishabh")
```
