```python
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal Name: {self.name}")


class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # calling parent constructor
        self.breed = breed
        print(f"Dog Breed: {self.breed}")


# Creating object
d = Dog("Tommy", "Labrador")
```
