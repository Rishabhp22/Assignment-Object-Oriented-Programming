ACCESS MODIFIERS (Core idea)

👉 Access modifiers decide:
➡️ “Who can access this variable or method?”

⚙️ Types in Python
🔹 1. Public

👉 Accessible from anywhere

```python
class Student:
    def __init__(self):
        self.name = "Rishabh"
```

✔ Access:

```python
s = Student()
print(s.name)   # works
```

2. Protected (_ single underscore)

👉 Meant for internal use (convention only)

```python
class Student:
    def __init__(self):
        self._marks = 90
```

✔ Access:

```python
print(s._marks)   # works (but not recommended)
```

🔹 3. Private (__ double underscore)

👉 Restricted access (name mangling)

```python
class Student:
    def __init__(self):
        self.__age = 20
```

❌ Direct access:

```python
print(s.__age)   # error
```

✔ Indirect:


```python
print(s._Student__age)
```
