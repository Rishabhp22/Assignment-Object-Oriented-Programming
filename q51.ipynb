Why Encapsulation is a Pillar of OOP

👉 Encapsulation is considered a pillar of OOP because it helps in protecting data and controlling access to it.

🔥 Key Reasons
1. Data Hiding

👉 Internal data is hidden from outside access
👉 Prevents direct modification

2. Controlled Access

👉 Access is provided through methods (getter/setter)
👉 Ensures safe data handling

3. Security

👉 Sensitive data cannot be easily accessed or changed

4. Maintainability

👉 Changes in internal code do not affect outside code

🧠 Example idea

👉 Bank account:

balance is private
accessed using methods
