What is __len__ ?

👉 __len__ is a special (dunder) method that defines what should be returned when we use len(object)

```python
class MyList:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)


obj = MyList([1, 2, 3, 4])
print(len(obj))
```
