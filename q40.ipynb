🔹 How to achieve abstraction in Python

👉 Abstraction is achieved using:

Abstract Base Classes (ABC)
@abstractmethod decorator
🔥 Steps
1. Import required module

```python
from abc import ABC, abstractmethod
```

2. Create abstract class

```python
class Animal(ABC):
```

3. Define abstract method

```python
@abstractmethod
def sound(self):
    pass
```

4. Create child class and implement method

```python
class Dog(Animal):
    def sound(self):
        print("Bark")
```
