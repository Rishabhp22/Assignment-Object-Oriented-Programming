Definition

👉 Polymorphism in OOP refers to the ability of a single object or function to take multiple forms depending on the context.

Same name → different behavior

Example:

+ → add numbers
+ → join strings

```python
print(2 + 3)        # 5
print("Hello " + "World")   # Hello World
```

Types of polymorphism
1. Compile-time (Method Overloading)

👉 Same function name, different parameters
(Python supports it in a limited way)

2. Runtime (Method Overriding) ⭐ IMPORTANT

👉 Child class changes parent method

```python
class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

d = Dog()
d.sound()
```
