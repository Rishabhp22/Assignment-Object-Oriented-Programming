What is __add__ ?

👉 __add__ is a special (dunder) method that defines how the + operator works for objects

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value


n1 = Number(10)
n2 = Number(20)

print(n1 + n2)
```
