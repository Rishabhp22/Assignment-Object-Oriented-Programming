```python
class Animal:
    def move(self):
        print("Animal moves")

class Dog(Animal):
    def move(self):
        print("Dog runs")

class Mammal:
    def reproduce(self):
        print("Giving birth to live young")

class DogMammal(Dog, Mammal):
    pass

d = DogMammal()
d.move()
d.reproduce()
```
