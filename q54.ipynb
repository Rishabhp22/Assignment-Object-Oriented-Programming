```python
class Mydecorator1:
    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        print("class Work start")
        result = self.func(*args, **kwargs)
        print("class work end")
        return result

class Mydecorator2:
    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        print("Home work start")
        result = self.func(*args, **kwargs)
        print("Home work end")
        return result

@Mydecorator1
@Mydecorator2
def work(a, b):
    result = a*b
    print(result)
    # return result

work(65, 65)
```
