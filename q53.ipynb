```python
class MyDecorator:
    def __init__(self, message):
        self.message = message

    def __call__(self, func):
        def wrapper(*args, **kwargs):
            print(f"{self.message} - Function: {func.__name__}")
            return func(*args, **kwargs)
        return wrapper

@MyDecorator("Executing")
def greet():
    print("Hello")

greet()
```
