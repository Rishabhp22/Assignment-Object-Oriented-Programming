👉 Inheritance is a mechanism in OOP where a child class acquires the properties (attributes) and methods of a parent class.

Purpose (important — most students miss this)
Code reuse
Avoid repetition
Build relationships between classes

```python
class Parent:
    pass

class Child(Parent):
    pass
```

```python
class Animal:
    def speak(self):
        print("Animal makes sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()   # inherited
d.bark()
```
